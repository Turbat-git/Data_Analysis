```python
%pip install pandas
import pandas as pd
df_solar = pd.read_csv('sgu-solar-installations-2011-to-present-and-totals.csv')
df_capacity = pd.read_csv('sgu-solar-capacity-2011-to-present-and-totals.csv')
df_population = pd.read_csv('Australian-Population-1981.csv')
df_state_pop = pd.read_csv('State-Population-1981.csv')
```

```python
df_population_erp = df_population[["cc", "Estimated Resident Population (ERP) ;  Australia ;"]]

df_december = df_population_erp[df_population_erp["cc"].str.contains("Dec")].reset_index(drop=True)

df_december = df_december.rename(columns={"cc": "Year", "Estimated Resident Population (ERP) ;  Australia ;": "Estimated Resident Population (000)"})
df_december["Year"] = df_december["Year"].str.replace("Dec-", "", regex=False)
df_december["Year"] = df_december["Year"].astype(int)
df_december = df_december.set_index("Year")
df_december.head(15)
```

```python
df_state = df_state_pop[df_state_pop["Year"].str.contains("Dec")].reset_index(drop=True)
df_state.columns = df_state.columns.str.strip() 
df_state.columns = df_state.columns.str.replace("Estimated Resident Population ; ", "", regex=False)
df_state.columns = df_state.columns.str.replace(" ;", "", regex=False)

male_df = df_state.filter(like="Male").copy()
female_df = df_state.filter(like="Female").copy()
male_df.columns = male_df.columns.str.replace("Male ", "", regex=False)
female_df.columns = female_df.columns.str.replace("Female ", "", regex=False)

df_state_total = male_df + female_df

df_state.head(10)
df_state = pd.concat([df_state, df_state_total], axis=1)
```
